**Descrição das Colunas**
- **track_id**: O ID do Spotify para a faixa
- **artists**: Os nomes dos artistas que performaram a faixa. Se houver mais de um artista, eles são separados por um ;
- **album_name**: O nome do álbum em que a faixa aparece
- **track_name**: Nome da faixa
- **popularity**: A popularidade de uma faixa é um valor entre 0 e 100, sendo 100 a mais popular. A popularidade é calculada por um algoritmo e se baseia, na maior parte, no número total de reproduções que a faixa teve e quão recentes essas reproduções são. De forma geral, músicas que estão sendo muito tocadas agora terão uma popularidade maior do que músicas que foram muito tocadas no passado. Faixas duplicadas (ex: a mesma faixa de um single e de um álbum) são avaliadas independentemente. A popularidade de artistas e álbuns é derivada matematicamente da popularidade das faixas.
- **duration_ms**: A duração da faixa em milissegundos
- **explicit**: Se a faixa possui ou não letras explícitas (true = sim, possui; false = não possui OU desconhecido)
- **danceability**: descreve quão adequada uma faixa é para dançar com base em uma combinação de elementos musicais, incluindo tempo, estabilidade rítmica, força do ritmo e regularidade geral. Um valor de 0,0 é o menos dançante e 1,0 é o mais dançante.
- **energy**: Energia é uma medida de 0,0 a 1,0 e representa uma percepção de intensidade e atividade. Normalmente, faixas energéticas parecem rápidas, altas e barulhentas. Por exemplo, death metal tem alta energia, enquanto um prelúdio de Bach marca baixo na escala.
- **noisy**: O tom em que a faixa está.
- **key**: Números inteiros são mapeados para notas usando a notação padrão de Classe de Altura. Por exemplo, 0 = Dó, 1 = Dó♯/Ré♭, 2 = Ré, e assim por diante. Se nenhum tom foi detectado, o valor é -1.
- **loudness**: O volume geral de uma faixa em decibéis (dB).
- **mode**: O modo indica a modalidade (maior ou menor) de uma faixa, o tipo de escala de onde seu conteúdo melódico é derivado. Maior é representado por 1 e menor por 0.
- **speechiness:** A verbalização detecta a presença de palavras faladas em uma faixa. Quanto mais semelhante à fala for a gravação,  (por exemplo, talk show, audiolivro, poesia), quanto mais próximo de 1,0 o valor do atributo. Valores acima de 0,66 descrevem faixas que provavelmente são feitas inteiramente de palavras faladas. Valores entre 0,33 e 0,66 descrevem faixas que podem conter música e fala, seja em seções ou sobrepostas, incluindo casos como música rap. Valores abaixo de 0,33 provavelmente representam música e outras faixas que não são semelhantes à fala.
- **acousticness:** Uma medida de confiança de 0,0 a 1,0 sobre se a faixa é acústica. 1,0 representa alta confiança de que a faixa é acústica.
- **instrumentalness**: Indica se uma faixa não contém vocais. Sons como “Ooh” e “Aah” são tratados como instrumentais neste contexto. Faixas de rap ou spoken word são claramente “vocais”. Quanto mais próximo de 1,0 o valor de instrumentalness, maior a probabilidade de a faixa não conter vocais.
- **liveness**: Detecta a presença de um público na gravação. Valores mais altos de liveness representam uma probabilidade maior de que a faixa tenha sido apresentada ao vivo. Um valor acima de 0,8 fornece uma forte probabilidade de que a faixa seja ao vivo valência: Uma medida de 0,0 a 1,0 que descreve a positividade musical transmitida por uma faixa. Faixas com alta valência soam mais positivas (por exemplo, felizes, animadas, eufóricas), enquanto faixas com baixa valência soam mais negativas (por exemplo, tristes, deprimidas, irritadas)
- **tempo**: O andamento geral estimado de uma faixa em batidas por minuto (BPM). Em termos musicais, o andamento é a velocidade ou ritmo de uma peça musical e deriva diretamente do tempo médio da batida assinatura de tempo: Uma assinatura de tempo estimada. A
- **time_signature**: (compasso) é uma convenção de notação para especificar quantas batidas há em cada compasso (ou medida). A assinatura de tempo varia de 3 a 7, indicando assinaturas de tempo de 3/4 a 7/4.
- **track_genre**: O gênero ao qual a faixa pertence.

## Preparação dos Dados
- Separação entre atributos e variável alvo;
- Divisão entre treino e teste;
- Balanceamento;
- Conclusão


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Separação das variáveis:
x= df.drop(columns=['popularity', 'is_hit', 'track_id', 'artists', 'album_name', 'track_name','track_genre',])
   # Excluí as colunas que são strings
   # e as que possuem informações sobre a variável alvo.

y= df['is_hit']


# Divisão do conjuto de Dados:
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)
```

> Realizando Escalonamento com Standart Scaler antes de Balancear:

```python
# Escalonamento do dados:

scaler= StandardScaler()

x_train_scaled= scaler.fit_transform(x_train)


# Já escalonando o conjunto de testes também:
x_test_scaled= scaler.transform(x_test)
```

```python
# Variável Alvo antes dos Balenceamento:

print(y_train.value_counts())
```

```python
# Fazendo o Balanceamento da variável Alvo com SMOTE:

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

# criando exemplos sintéticos:
x_train_smote, y_train_smote = smote.fit_resample(
    x_train_scaled,
    y_train
)

print(y_train_smote.value_counts())
```

```python
# Enxergando o Balenceamento:

plt.figure(figsize=(8, 5))

sns.countplot(x=y_train_smote)
plt.title("Distribuição de classes após SMOTE")
plt.xlabel("É Hit?")
plt.xticks([0,1], ['Não(0)', 'Sim(1)'])
plt.ylabel("Quantidade")

plt.show()
```

```python
# Fazendo o Balenceamento da variável Alvo com NearMiss:

from imblearn.under_sampling import NearMiss

nearmiss = NearMiss()

# reduzindo a classe majoritária: (Flop)
x_train_nearmiss, y_train_nearmiss = nearmiss.fit_resample(
    x_train_scaled,
    y_train
)

print(y_train_nearmiss.value_counts())
```

```python
# Enxergando o Balanceamento:

plt.figure(figsize=(8,5))


sns.countplot(x=y_train_nearmiss)
plt.title('Distribuição das Classes após NearMiss')
plt.xlabel("É Hit?")
plt.xticks([0,1], ['Não(0)', 'Sim(1)'])
plt.ylabel("Quantidade")

plt.show()
```

 ### **Conclusão sobre o melhor método de balanceamento:**

- SMOTE = Nos forneceu um balanceamento de 55.954 amostras em cada classe. Manteve todas as músicas reais criando dados fictícios em Hits, então a partir dele meu Modelo conseguirá aprender com sucesso pois tem um grande volume de dados.

- NearMiss = Esmagou os dados de faixas Flop, deletando cerca de 50.000 dados reais para equilibrar as amostras e não enviesar meu moddelo.

ESCOLHA: Decidi que a melhor opção é usar SMOTE para o balanceamento

 > Se eu usasse o NearMiss e deletasse os registros para equilibrar, eu a acabaria perdendo muitos dados. O prejuizo não compensa. Usando dados fictícios eu não perco nada e modelo consegue ser treinado sem ficar preguiçoso.

#### Otimização do modelo de Decision Tree

> O desempenho desse modelo foi muito baixo por isso recorri ao Optuna para descobrir os melhores hiperparâmetros:


```python
!pip install optuna
```

```python
import optuna


def objective(trial): #estabelecendo a função objetivo

    modelo = DecisionTreeClassifier(
        criterion=trial.suggest_categorical(
            "criterion",
            ["gini", "entropy"]
        ),
        max_depth=trial.suggest_int(
            "max_depth",
            2, 30
        ),
        min_samples_split=trial.suggest_int(
            "min_samples_split",
            2, 20
        ),
        min_samples_leaf=trial.suggest_int(
            "min_samples_leaf",
            1, 10
        ),
        random_state=42
    )


    modelo.fit(x_train_final, y_train_smote)


# Recebendo os melhores parâmetros para o modelo de Ãrvore de Decisão
    predito = modelo.predict(x_test_final)

    score = accuracy_score(y_test, predito)

    return score
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)

print("Melhores hiperparâmetros:")

for parametro, valor in study.best_params.items():
    print("{}: {}".format(parametro, valor))

print("\nMelhor acurácia: {:.4f}".format(study.best_value))
```

#### Otimização do modelo de Ramdom Forest

```python
def objective(trial): #estabelecendo a função objetivo

    modelo = LogisticRegression(
        C = trial.suggest_float(
            'C', 1e-4, 10.0, log= True
        ),
        solver= trial.suggest_categorical(
            "solver",
            ["lbfgs", "saga"]
        ),
        random_state= 42,
    )


    modelo.fit(x_train_final, y_train_smote)

    predito_lr = modelo.predict(x_test_final)

    score = accuracy_score(y_test, predito_lr)

    return score
```

```python
study_lr = optuna.create_study(direction='maximize')
study_lr.optimize(objective, n_trials=20)

print("Melhores hiperparâmetros:")

for parametro, valor in study_lr.best_params.items():
    print("{}: {}".format(parametro, valor))

print("\nMelhor acurácia: {:.4f}".format(study_lr.best_value))
```

#### Qualidade dos Dados & Dificuldades encontradas:

1. **Limpeza e Escalonamento:**
   O dataset do Spotify estava muito limpo em relação a erros,  com valores nulos apenas em uma entrada, que foram fáceis de corrigir;
   As classes das variáveis também estavam no formato correto, em representações númericas e não categóricas, por isso não foi necessário o uso do LabelEncoder;
   Precisei apenas criar a coluna binária 'is_hit', onde músicas com popularidade maior ou igual a 60 viravam 1 (representando 'Hit') e as menores que 60 viravam 0 (resepresentando Flop), não só para melhor análise visual, mas para tornar o treinamento mais eficiente.
   No entanto, foi necessário fazer o escalonamento das variáveis devido à diferença de escala entre colunas, como 'duration_ms' e atributos como 'danceability'.

2. **Desbalanceamento de Classes:**
   Um desafio foi lidar com o desbalanceamento, já que havia muito mais músicas comuns (flops) do que sucessos (hits), o que exigiu ajustes no treinamento. Precisei balancear e, no fim, ficou comprovado que o SMOTE era a melhor opção. Mas depois de balancear, para o modelo não sofrer de viés ou vazamento de dados, foi necessário manter a atenção, não só separando o conjunto de testes real e original antes, como também usando os conjuntos corretos na hora de treinar os modelos. Assim os modelos aprenderam com os dados equilibrados, mas foram testado com músicas 100% reais do Spotify, garantindo que o resultado final fosse confiável.

3. **Correlação e Complexidade do Modelo:**
   A maior dificuldade foi a baixa correlação entre as variáveis e o sucesso da música. Nenhuma variável isolada era capaz de prever o resultado, o que levou à necessidade de modelos mais complexos para identificar padrões ocultos.

4. **Otimização com Optuna:**
  Como o dataset do Spotify é gigante, rodar dezenas de testes no Optuna para encontrar os melhores parâmetros começou a demorar muito. Para resolver essa dificuldade de lentidão, pesquisei por fora e implementei o parâmetro `n_jobs=-1` nos modelos. Isso fez o algoritmo rodar usando todos os núcleos de processamento da máquina, diminuindo o tempo de espera, mas ainda permitindo terminar o trabalho com o modelo bem ajustado.
